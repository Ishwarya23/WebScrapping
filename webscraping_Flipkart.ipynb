```python
import requests
from bs4 import BeautifulSoup
import urllib.request
```

```python
url = "https://www.flipkart.com/laptops/~cs-8ry1xi3u60/pr?sid=6bo%2Cb5g&collection-tab-name=intel%2011th%20gen&param=01&otracker=clp_banner_1_1.banner.BANNER_la-ops-40020-store_SNI4QTRS197A"
response = requests.get(url)
response
```

```python
htmlcontent = response.content
htmlcontent
```

```python
soup = BeautifulSoup(htmlcontent,"html.parser")
print (soup.prettify)
```

```python
laptopName = []
prices = []
rating = []
name = soup.find('div',attrs={'class':'_4rR01T'})
name
```

```python
laptopName = []
prices = []
rating = []
for i in soup.find_all('a',href=True,attrs={'class':'_1fQZEK'}):
    name = i.find('div',attrs={'class':'_4rR01T'})
    price = i.find('div',attrs={'class':'_30jeq3 _1_WHN1'})
    laptopName.append(name.text)
    prices.append(price.text)
    try:
        rate = i.find('div',attrs={'class':'_3LWZlK'})
        rating.append(rate.text) 
    except:
        rating.append("None")   
print (len(laptopName))
print (len(prices))
print (len(rating))
```
